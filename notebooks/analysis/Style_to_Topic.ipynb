## Get topic proportions by class style


```python
by_style_mean = df_topics[['class_style', 'Balance and Alignment', 'Stress Relief/Breath Techniques', 'Deep Relaxation/Mind Practice', 'Stretch and Release', 'Core and Strength']].groupby(['class_style'], as_index=False).mean()
#by_style_mean = by_style_mean.round(2)
#by_style_mean.loc[:,'Row_Total'] = by_style_mean.sum(numeric_only=True, axis=1) #check if the rows sum to 1
by_style_mean
```

```python
_, ax = plt.subplots(figsize=(15, 10))
by_style_mean.plot.barh(x='class_style',stacked=True, alpha=0.8, ax=ax)
ax.set_ylabel('Class Style', size=12)
ax.set_xlabel('% of belonging to a topic', size=12)
ax.legend(fontsize="15", bbox_to_anchor = (1.1, 0.5), loc='center right', fancybox=True, shadow=True)
plt.show()
```
